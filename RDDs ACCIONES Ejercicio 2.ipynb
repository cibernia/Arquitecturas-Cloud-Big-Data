```
# subir archivo sensores.txt que contiene las medidas de humedad en formato separado por comas
#Archivo->Upload datos en DBFS...
file = 'dbfs:/FileStore/shared_uploads/ciberniainfo@gmail.com/SENSORES.txt'
medidas = sc.textFile(file)
```

```
separar_medidas = medidas.flatMap(lambda elto: elto.split(','))
separar_medidas.take(10)
```

```
#pasamos los valores que están en formato texto a formato numérico
final_medidas = separar_medidas.map(float)
final_medidas.take(10)
```

```
#calculamos la media, minimo y maximo de las medidas de humedad
suma_total_medidas = final_medidas.reduce(lambda e1,e2: e1 + e2)
numero_medidas = final_medidas.count()
print ('numero de medidas realizadas', numero_medidas)
print ('el valor maximo es:         ', final_medidas.max())
print ('el valor mínimo es:         ', final_medidas.min())
print ('valor medio de medida:      ', round(suma_total_medidas/numero_medidas,1))# ,1 indica el numero de decimales
```
