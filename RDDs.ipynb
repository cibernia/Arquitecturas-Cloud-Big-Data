```
#.map
numeros = sc.parallelize ([1,2,3,4,5])
num3 = numeros.map(lambda elemento: 3*elemento)# especifico que operación hacer a cada elemento con lambda, num3 es un nuevo RDD, la funcion map el tamaño del RDD de entrada es el mismo que el de salida.
num3.collect()
```

```
#.map
palabras = sc.parallelize(['HOLA', 'Que', 'TAL', 'Bien'])
pal_minus = palabras.map(lambda elemento: elemento.lower())

pal_minus.collect()
```

```
#.filter, el RDD de salida es igual o menor que el RDD de entrada
numeros = sc.parallelize([1,2,3,4,5])
pares_rdd = numeros.filter(lambda elemento: elemento%2==0)
                           
pares_rdd.collect()
```

```
#filter
log = sc.parallelize(['E: e21', 'W: w12', 'W: w13', 'E: e45'])
errors = log.filter(lambda elemento: elemento[0]=='E')# que su primer caracter empiece por 'E'

errors.collect()
```

```
#crear un nuevo RDD que contenga 12 en su elemenento
log = sc.parallelize(['E: e21', 'W: w12', 'W: w13', 'E: e45'])
errors = log.filter(lambda elemento: elemento[-2]=='1' and elemento[-1]=='2')

errors.collect()
```
