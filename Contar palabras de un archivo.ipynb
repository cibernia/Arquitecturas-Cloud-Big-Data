```
# subir achivo quijote.txt en menú Archivo->Upload datos en el DBFS... y copiar la ruta del archivo
# mi resultado: dbfs:/FileStore/shared_uploads/ciberniainfo@gmail.com/quijote.txt

file = 'dbfs:/FileStore/shared_uploads/ciberniainfo@gmail.com/quijote.txt' # ojo poner entre comillas '  '

lineas = sc.textFile(file)
print ('numweo de lineas -> ', lineas.count()) # no utilizar collect porq el tamaño del archivo ya es grande
print('primeras cinco lineas ->' ,lineas.take(5)) # entrega las primeras 5 lineas en formato de lista, trae también las neneas en blanco ''
```

```
palabras_linea = lineas.map(lambda elemento: elemento.split())
palabras_linea.take(5)
```

```
num_palabras_linea = palabras_linea.map(lambda elemento: len(elemento))
num_palabras_linea.take(5)
```

```
num_palabras_linea.reduce(lambda elemento1, elemento2: elemento1+elemento2)
```
